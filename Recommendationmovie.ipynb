```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import seaborn as sns
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies_df = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits_column_renamed = credits.rename(index=str, columns= {"movie_id": "id"})
```

```python
credits_column_renamed
```

```python
movies_df_merge = movies_df.merge(credits_column_renamed, on= "id")
movies_df_merge.head()
```

```python
movies_df_merge.shape
```

```python
movies_cleaned_df = movies_df_merge.drop(columns = ['homepage','title_x','title_y','status'])
```

```python
movies_cleaned_df.shape
```

```python
movies_cleaned_df.info()
```

```python
features_with_na = [features for features in movies_cleaned_df.columns if movies_cleaned_df[features].isnull().sum()>1]
```

```python
features_with_na
```

```python
movies_cleaned_df[features_with_na].isnull().sum()
```

```python
features_with_na.remove('runtime')
```

```python
features_with_na
```

```python
def replace_cat_feature(movies_cleaned_df,features_with_na):
    data = movies_cleaned_df.copy()
    data[features_with_na]=data[features_with_na].fillna('Missing')
    return data
```

```python
movies_cleaned_df = replace_cat_feature(movies_cleaned_df,features_with_na)
```

```python
movies_cleaned_df[features_with_na].isnull().sum()
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df['runtime'].mean()
```

```python
movies_cleaned_df['runtime'].fillna(106.8, inplace=True)
```

```python
movies_cleaned_df['runtime'].isnull().sum()
```

```python
''' Using Weighted average for each movies average rating.
                        
                    W = Rv + Cm
                       ---------
                         v + m
        Where 
        W = Weighted Rating
        R = average for the movie as a number 0 to 10 (mean) = (Rating)
        v = Number of votes for the movie= (votes)
        m = minimum votes required to be listed in the top 250( Currently 300)
        C = The meam votes across the Whole report( currently 6.9)'''
```

```python
v = movies_cleaned_df['vote_count']
R = movies_cleaned_df['vote_average']
C = movies_cleaned_df['vote_average'].mean()
m = movies_cleaned_df['vote_count'].quantile(0.70)
```

```python
movies_cleaned_df['weighted_average'] = ((R*v) + (C*m)) / (v + m)
```

```python
movies_cleaned_df['weighted_average']
```

```python
movie_sorted_ranking = movies_cleaned_df.sort_values('weighted_average', ascending = False)
```

```python
movie_sorted_ranking[['original_title', 'vote_count','vote_average','weighted_average','popularity']]
```

```python
weight_average = movie_sorted_ranking.sort_values('weighted_average', ascending= False)
plt.figure(figsize = (12,6))
axis1 = sns.barplot(x=weight_average['weighted_average'].head(10), y=weight_average['original_title'].head(10))
plt.xlim(4, 10)
plt.title('Best movies by average votes', weight = 'bold')
plt.xlabel('Weighted average score', weight='bold')
plt.ylabel('Movie_tile', weight = 'bold')
```

```python
popularity = movie_sorted_ranking.sort_values('popularity', ascending = False)
```

```python
popularity.head()
```

```python

plt.figure(figsize = (12,6))
axis1 = sns.barplot(x=popularity['popularity'].head(10), y=popularity['original_title'].head(10), data= popularity)
plt.xlim(4, 10)
plt.title('Best movies by popularity', weight = 'bold')
plt.xlabel('popularity', weight='bold')
plt.ylabel('Movie_tile', weight = 'bold')
```

```python
#Recommendation based on scaled weighted average and popularity , 50% weightage is given to both
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(movies_cleaned_df[['weighted_average','popularity']])
```

```python
movie_scaled_df
```

```python
movie_normalized_df = pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
movies_cleaned_df[['normalized_weight_average','normalized_popularity']] = movie_normalized_df
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score'] = movies_cleaned_df['normalized_weight_average'] * 0.5 + movies_cleaned_df['normalized_popularity'] * 0.5
```

```python
movies_cleaned_df.head()
```

```python
movies_score_df = movies_cleaned_df.sort_values('score', ascending = False)
```
