```python
'''In exploratory data analysis we have understood the relation between respective features,In this feature engineering section we will
be dealing with following things

1.Missing Values
2.Temporal Variables
3.Categorical Variables
4.Remove rare labels
5.Standardize the values of the variables, to the same range'''
```

```python
'''Feature engineering has to be applied in both train and test data,in order to avoid data leakage.'''

# Missing values of categorical features.
```

```python
features_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes == 'O']
```

```python
features_nan
```

```python
for feature in features_nan:
    print('{}:{}% missing values'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
#Replace missing value with a new label to manage missing values
```

```python
def replace_cat_feature(dataset,features_nan):
    data = dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data
```

```python
dataset = replace_cat_feature(dataset,features_nan)
```

```python
dataset
```

```python
dataset[features_nan].isnull().sum()
```

```python
#As we have seen categorical variables were having null values , we have replaced null values with a new label.
```

```python
# Now let's check for numerical variables that contains missing values
```

```python
numerical_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes != 'O']
```

```python
numerical_with_nan
```

```python
for feature in numerical_with_nan:
    print("{}:{}% missing value".format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
# In exploratory data analysis we have seen there are lot of out liers , so when we are having
#lot of out liers we need to fill nan values with median . filling with mean doesn't work here.
```

```python
for feature in numerical_with_nan:
    median_value = dataset[feature].median()
    #create a new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace = True)
```

```python
dataset[numerical_with_nan].isnull().sum()
```

```python
dataset.head()
```

```python
#Temporal variable(date and time variable management)
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature] = dataset['YrSold'] - dataset[feature]
```

```python
dataset.head()
```

```python
''' We are observing above code in which we are converting year features into number of years
so in place 1962 it will be some 58 years. So in this way years are converted to numbers.Here basically YrSold feature is subtracted
from other year features in order to convert temporal variable to numbers'''
```

```python
dataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
# In exploratory data analysis we have seen YrSold was reducing so this feature is not considered.
```

```python
#NUMERICAL VARIABLES
'''As we have seen numerical variables are not obeying gussian distribution, also those are skewed data, so it will 
be better to use log normal distribution. Here we are not selecting numerical variables containing 0 value.
Which are skewed only those will be considered here'''
```

```python
num_feature = ['LotFrontage','LotArea','1stFlrSF','GrLivArea']
```

```python
num_feature
```

```python
for feature in num_feature:
    dataset[feature] = np.log(dataset[feature])
```

```python
dataset.head()
```

```python
#Handling rare categorical feature
```

```python
'''We will be removing categorical variables that are present less than 1% of the observations'''
```

```python
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtype == 'O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp = dataset.groupby(feature)['SaleCondition'].count()/len(dataset)
    temp_df = temp[temp > 0.01].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head()
```

```python
#Here in above code 1st% of feature is taken, then a new is made where index value will be stored.
#If dataset index value is stored & found that feature value that feature value will be replaced by a new label.
#Here new labl is rare_var
```

```python
for feature in categorical_features:
    labels_ordered = dataset.groupby([feature])['GrLivArea'].mean().sort_values().index
    labels_ordered = {k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature] = dataset[feature].map(labels_ordered)
```
